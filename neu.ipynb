```python
import matplotlib.pyplot as plt
import numpy as np

def create_simple_nn_diagram(input_size, hidden_sizes, output_size, figsize=(10, 6)):
    """
    Create a simple neural network diagram with clearly visible layers.
    """
    fig, ax = plt.subplots(figsize=figsize)
    
    # Colors for each layer
    colors = ['#3498db', '#2ecc71', '#e74c3c']  # Blue, Green, Red
    
    # Layer names and sizes
    layer_names = ['Input Layer', 'Hidden Layer', 'Output Layer']
    layer_sizes = [input_size, hidden_sizes[0], output_size]
    
    # Position parameters
    x_positions = [1, 3, 5]
    max_height = 4
    box_width = 1.2
    
    # Draw each layer as a rectangle
    for i, (name, size, x_pos, color) in enumerate(zip(layer_names, layer_sizes, x_positions, colors)):
        # Calculate height based on relative size but with a maximum
        relative_size = size / max(layer_sizes)
        display_size = min(relative_size, 0.8) * max_height if i > 0 else 0.8 * max_height
        
        # Draw rectangle
        rect = plt.Rectangle((x_pos - box_width/2, 2 - display_size/2), 
                           box_width, display_size, 
                           facecolor=color, alpha=0.7, edgecolor='black')
        ax.add_patch(rect)
        
        # Add layer name and size
        if size < 100:
            size_text = f"{size} neurons"
        else:
            size_text = f"{size:,} neurons"
            
        plt.text(x_pos, 2 + display_size/2 + 0.3, name, ha='center', fontsize=12)
        plt.text(x_pos, 2 - display_size/2 - 0.3, size_text, ha='center', fontsize=10)
        
        # Add icons or symbols for special layers
        if i == 1:  # Hidden layer
            plt.text(x_pos, 2, "ReLU", ha='center', fontsize=10, color='white', weight='bold')
            plt.text(x_pos, 1.0, "Dropout (0.3)", ha='center', fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.3", fc='#f39c12', ec="black", alpha=0.7))
    
    # Draw arrows between layers
    for i in range(len(x_positions) - 1):
        arrow_start = x_positions[i] + box_width/2
        arrow_end = x_positions[i+1] - box_width/2
        plt.arrow(arrow_start, 2, arrow_end - arrow_start - 0.1, 0, 
                 head_width=0.1, head_length=0.1, fc='black', ec='black', length_includes_head=True)
    
    # Add title and details
    plt.title('Feedforward Neural Network Architecture', fontsize=14, pad=20)
    
    # Add model specs as text
    spec_text = f"Input: {input_size} → Hidden: {hidden_sizes[0]} → Output: {output_size}\n"
    spec_text += "Activation: ReLU | Loss: CrossEntropy | Optimizer: Adam (lr=1e-4)"
    plt.figtext(0.5, 0.05, spec_text, ha='center', fontsize=11, style='italic')
    
    # Set limits and remove axes
    plt.xlim(0, 6)
    plt.ylim(0, 4)
    plt.axis('off')
    
    plt.tight_layout()
    return fig

# Create the visualization
input_size = 10000
hidden_sizes = [128]
output_size = 10

fig = create_simple_nn_diagram(input_size, hidden_sizes, output_size)
plt.savefig('simple_nn_architecture.png', dpi=300, bbox_inches='tight')
plt.show()

# Confirm layer creation
print("Neural Network Diagram created with THREE clearly visible layers:")
print(f"1. Input Layer (Blue): {input_size:,} neurons")
print(f"2. Hidden Layer (Green): {hidden_sizes[0]} neurons with ReLU activation and Dropout(0.3)")
print(f"3. Output Layer (Red): {output_size} neurons")
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

def plot_nn_scaled_ratio(real_sizes, max_display_neurons=12, min_display_neurons=2):
    G = nx.DiGraph()

    max_real_size = max(real_sizes)
    
    # Dynamically decide display neurons
    display_sizes = []
    for size in real_sizes:
        if size == max_real_size:
            display = max_display_neurons
        else:
            ratio = size / max_real_size
            display = int(np.clip(ratio * max_display_neurons, min_display_neurons, max_display_neurons))
        display_sizes.append(display)

    # Create nodes
    layer_nodes = []
    for idx, num_nodes in enumerate(display_sizes):
        nodes = []
        for i in range(num_nodes):
            node_name = f"L{idx}_N{i}"
            G.add_node(node_name, layer=idx)
            nodes.append(node_name)
        layer_nodes.append(nodes)

    # Connect nodes layer by layer
    for (layer1, layer2) in zip(layer_nodes[:-1], layer_nodes[1:]):
        for node1 in layer1:
            for node2 in layer2:
                G.add_edge(node1, node2)

    # Layout
    pos = {}
    for layer_idx, nodes in enumerate(layer_nodes):
        x = layer_idx * 2
        y_spacing = 1.5
        offset = -(len(nodes) - 1) / 2
        for i, node in enumerate(nodes):
            pos[node] = (x, y_spacing * (i + offset))

    fig, ax = plt.subplots(figsize=(14, 8))
    nx.draw(G, pos, with_labels=False, node_size=300, node_color='skyblue', edge_color='gray', ax=ax)

    # Add real neuron counts under each layer
    for idx, x_pos in enumerate(range(0, len(layer_nodes)*2, 2)):
        ax.text(x_pos, -5, f"{real_sizes[idx]} neurons", ha='center', fontsize=12, color='black')

    plt.title("Neural Network Architecture for mean pooling model", fontsize=16)
    plt.axis('off')
    plt.show()

# Your real model sizes
real_layer_sizes = [300, 265, 128, 64, 10]

plot_nn_scaled_ratio(real_layer_sizes)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

def plot_nn_layers(layers, max_display_neurons=12, min_display_neurons=2):
    G = nx.DiGraph()

    max_real_size = max([layer['size'] for layer in layers if layer['size'] is not None])

    display_sizes = []
    for layer in layers:
        if layer['size'] is None:
            display_sizes.append(1)  # Dropout layer: draw 1 node
        else:
            if layer['size'] == max_real_size:
                display = max_display_neurons
            else:
                ratio = layer['size'] / max_real_size
                display = int(np.clip(ratio * max_display_neurons, min_display_neurons, max_display_neurons))
            display_sizes.append(display)

    # Create nodes
    layer_nodes = []
    for idx, num_nodes in enumerate(display_sizes):
        nodes = []
        for i in range(num_nodes):
            node_name = f"L{idx}_N{i}"
            G.add_node(node_name, layer=idx)
            nodes.append(node_name)
        layer_nodes.append(nodes)

    # Connect nodes layer by layer
    for (layer1, layer2) in zip(layer_nodes[:-1], layer_nodes[1:]):
        for node1 in layer1:
            for node2 in layer2:
                G.add_edge(node1, node2)

    # Layout
    pos = {}
    for layer_idx, nodes in enumerate(layer_nodes):
        x = layer_idx * 2
        y_spacing = 1.5
        offset = -(len(nodes) - 1) / 2
        for i, node in enumerate(nodes):
            pos[node] = (x, y_spacing * (i + offset))

    fig, ax = plt.subplots(figsize=(16, 10))

    # Color nodes differently based on layer type
    node_colors = []
    for idx, nodes in enumerate(layer_nodes):
        layer_type = layers[idx]['type']
        if layer_type == 'Dense':
            node_colors.extend(['skyblue'] * len(nodes))
        elif layer_type == 'Dropout':
            node_colors.extend(['lightcoral'] * len(nodes))
        elif layer_type == 'Input':
            node_colors.extend(['lightgreen'] * len(nodes))
        elif layer_type == 'Output':
            node_colors.extend(['gold'] * len(nodes))
        else:
            node_colors.extend(['lightgray'] * len(nodes))

    nx.draw(G, pos, with_labels=False, node_size=300, node_color=node_colors, edge_color='gray', ax=ax)

    # Add layer labels
    for idx, x_pos in enumerate(range(0, len(layer_nodes)*2, 2)):
        layer = layers[idx]
        label = f"{layer['type']}"
        if layer.get('activation'):
            label += f"\n{layer['activation']}"
        if layer.get('dropout'):
            label += f"\nDropout {layer['dropout']}"
        
        ax.text(x_pos, -7, label, ha='center', fontsize=10, color='black')
        if layer['size']:
            ax.text(x_pos, -9, f"{layer['size']} neurons", ha='center', fontsize=9, color='black')

    plt.title("Neural Network Architecture simple RNN", fontsize=16)
    plt.axis('off')
    plt.show()

layers = [
    {'type': 'Embedding', 'size': None, 'activation': 'output_dim=64'},
    {'type': 'SimpleRNN', 'size': 64, 'activation': 'tanh + L2(0.001)'},
    {'type': 'BatchNormalization', 'size': None},
    {'type': 'Dropout', 'size': None, 'dropout': 0.5},
    {'type': 'Dense', 'size': 48, 'activation': 'ReLU + L2(0.001)'},
    {'type': 'Dropout', 'size': None, 'dropout': 0.4},
    {'type': 'Output', 'size': 10, 'activation': 'Softmax'}
]




plot_nn_layers(layers)
```
